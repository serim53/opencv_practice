```
import cv2
import matplotlib.pyplot as plt

image_1 = cv2.imread('pattern (3).jpg')
image_2 = cv2.imread('bag.jpg')

result = cv2.add(image_1, image_2)  #cv2.add를 이용할 때는 두 사진의 픽셀의 크기가 같아야함.
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()

result = image_1 + image_2
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```
