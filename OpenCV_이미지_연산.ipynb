OpenCV를 활용한 이미지 크기 및 픽셀 정보 확인

```
import cv2
image = cv2.imread('dog.jpg')

# 픽셀 수 및 이미지 크기 확인
print(image.shape)  # (높이, 너비, )
print(image.size)

# 이미지 Numpy 객체의 특정 픽셀을 가리킵니다.
px = image[100, 100]    # 범위를 넘어가는 픽셀을 가리키면 outofbound 오류가 남

# B, G, R 순서로 출력됩니다.
# (단, Gray Scale인 경우에는 B, G, R로 구분되지 않습니다.)
print(px)

# R 값만 출력하기
print(px[2])
```

OpenCV를 활용한 특정 범위 픽셀 변경

```
import cv2
import matplotlib.pyplot as plt
import time

image = cv2.imread('dog.jpg')
'''
start_time = time.time()
# (1)개별적인 픽셀을 일일이 접근해서 그 값을 바꿔주는 방법
for i in range(0, 100):
    for j in range(0, 100):
        image[i, j] = [255, 255, 255]
# 0부터 99까지 모든 픽셀이 하얀색으로 바뀜, 시간 오래걸림
print("--- %s seconds ---" % (time.time() - start_time))
'''
start_time = time.time()
# (2)범위를 지정한 다음 그 값을 동일한 어떠한 값으로 채워넣음
image[0:100, 0:100] = [0, 0, 0] 
print("--- %s seconds ---" % (time.time() - start_time))
#처리속도 빠름
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

OpenCV를 활용한 ROI(Region of Interest: 관심 있는 영역) 추출

```
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('dog.jpg')

# Numpy Slicing: ROI 처리 가능
roi = image[250:450, 150:350]

# ROI 단위로 이미지 복사하기
image[0:200, 0:200] = roi

# 이미지 범위공간을 다르게 설정하면 오류 발생

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

OpenCV를 활용한 픽셀별 색상 다루기

```
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('dog.jpg')
image[:, :, 2] = 0  # 빨간색들을 모두 0으로 만듬. 초록색과 파란색에 대한 정보만 남음

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
